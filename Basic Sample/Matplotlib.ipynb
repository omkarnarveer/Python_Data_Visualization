# Write a Program to illustrate Scatter Plot

```python
# Import matplotlib
import matplotlib.pyplot as plt

# Set plot space as inline for inline plots and qt for external plots
%matplotlib inline

# Declaring  X and y values
x =[5, 7, 8, 7, 2, 17, 2, 9,
    4, 11, 12, 9, 6] 
  
y =[99, 86, 87, 88, 100, 86, 
    103, 87, 94, 78, 77, 85, 86]

# Set the figure size in inches
plt.figure(figsize=(10,6))

plt.scatter(x, y, label = "Point", c = "blue")

# Set x and y axes labels
plt.xlabel('X Values')
plt.ylabel('Y Values')

plt.title('Scatter Plot')
plt.legend()
plt.show()
```

# Write a Program to illustrate Box Plot

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
  
  
# Creating dataset
np.random.seed(10)
data = np.random.normal(100, 20, 200)
  
fig = plt.figure(figsize =(10, 7))
  
# Creating plot
plt.boxplot(data)
  
# show plot
plt.show()
```

# Write a Program to illustrate Pie Chart

```python
import matplotlib.pyplot as plt
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
# Plot
plt.pie(sizes, labels=labels, 
        autopct='%1.2f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

# Write a program to illustrate Stacked Bar Chart

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np
  
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
  
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

# Write a program to illustrate Bar Plot

```python

import matplotlib.pyplot as plt  
   
# Creating the dataset 
data = {'DBMS':20, 'CNS':15, 'SPOS':30,  
        'IOT':35} 
courses = list(data.keys()) 
values = list(data.values()) 
   

fig = plt.figure(figsize = (10, 6)) 
  
# creating the bar plot 
plt.bar(courses, values, color ='green',  
        width = 0.4) 
  
plt.xlabel("Courses offered") 
plt.ylabel("No. of students enrolled") 
plt.title("Students enrolled in different courses") 
plt.show() 
```
